```python
import tweepy
import pandas as pd
import time
import numpy as np
import matplotlib.pyplot as plt
```

For getting the secret API keys please refer https://developer.twitter.com/en/docs

```python
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
top_hashtag = "#FarmersAppealTotalRepeal"
date_since = "2020-12-30"
tweets = tweepy.Cursor(api.search,
                       q=top_hashtag+" -filter:retweets",
                       since=date_since,wait_on_rate_limit=True).items(20000)
```

# Collecting the top 20000 tweets

```python
counter=0

data = []
for tweet in tweets:
    data.append(tweet)
    counter+=1
    
    print(counter, len(data))    
    if len(data) > 20000:
        break
        
    if counter > 1000:
        time.sleep(15 * 60)
        counter=0
    
```

```python
import pickle

with open('data_tweets.pickle', 'wb') as handle:
    pickle.dump(data, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
